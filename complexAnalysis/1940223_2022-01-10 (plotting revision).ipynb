```python
from matplotlib.pyplot import *
from numpy import sin, cos, linspace
ax = axes(projection = '3d')
x = linspace(-10, 10, 100) # The arguments are: lower bound, upper bound, step size... it creates an array accordingly
y = sin(x)
z = cos(x)
ax.plot3D(x, y, z, 'red')
title("What is this???", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
xvalues = np.linspace(-np.pi, np.pi) # Creates a vector from the defined range
plt.suptitle('Polynomial Functions') # Title over all subplots

plt.subplot(2, 2, 1)
yvalues = xvalues**1
plt.plot(xvalues, yvalues, lw=2, color='red', label='y = x')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
yvalues = xvalues**2
plt.plot(xvalues, yvalues, lw=2, color='blue', label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 3)
yvalues = xvalues**3
plt.plot(xvalues, yvalues, lw=2, color='green', label='y = x^3')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 4)
yvalues = xvalues**4
plt.plot(xvalues, yvalues, lw=2, color='purple', label='y = x^4')
plt.xlabel('x')
plt.ylabel('y')

# The following axis drawing functions only apply for the just preceding plot.
plt.axhline(lw=0.5, color='purple')
plt.axvline(lw=0.5, color='purple')
plt.legend()

plt.tight_layout()
# Must be given after the subplots. Ensures that there is sufficient space between the subplots

None
```
