We will learn to solve a differential equation of the form

$f^n(x)+f^{n-1}(x)+...+f'(x)+f(x)=g(x)$

where g(x) is a known function of x, and f(x) is a yet unknown i.e. unsolved function of x. In the following examples with the order two, note that u0 is the list of values of f^n(0) for all n, i.e. u0 = [f(0),  f'(0), f''(0)...]. xs is the array of x values. In dy_dx, the function returning the differntials of y and y', we have that u[1] represents y' and u[0] represents y. The start with the inital values present in u0, but the odeint function uses these inital results to find the differentials y and y' for all x values present in xs. us contains a list of lists, each individual list containing the values of y and y' corresponding to the x values in xs. To access only the y values, we assign us[:, 0] to ys. Hence, ys contains all the y values corresponding to the x values in xs, for the given differential equation.

$y''+2y'+2y=cos(2x)$, with initial conditions $y(0)=0, y'(0)=0$

```python
from numpy import linspace, cos
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u, x):
    return [u[1], cos(2*x) - 2*u[1] - 2*u[0]]
u0 = [0, 0]
xs = linspace(0, 10, 200)
us = odeint(du_dx, u0, xs)
ys = us[:, 0]
plt.plot(xs, ys)
plt.title("Solution of $y''+2y'+2y=cos(2x)$")
plt.xlabel("x")
plt.ylabel("y")
None
```

$y''-y'+9y=0$, with initial conditions $y(0)=1, y'(0)=2$

```python
from numpy import linspace, cos
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u, x):
    return [u[1], u[1] - 9*u[0]]
u0 = [0, 1]
xs = linspace(0, 10, 200)
us = odeint(du_dx, u0, xs)
ys = us[:, 0]
plt.plot(xs, ys)
plt.title("Solution of $y''-y'+9y=0$")
plt.xlabel("x")
plt.ylabel("y")
None
```

# A closer look at the process...

Obtain the solution for the differential equation:

$\frac{d^2y}{dt^2}+3\frac{dy}{dt}-5y=0$, where $y(0) = 0, y'(0) = 5$

at time points t are 5, 10 and 15. Represent them by scatterplot.

```python
from numpy import linspace, cos
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def dy_dx(y, t):
    return [y[1], - 3*y[1] + 5*y[0]]
initialValues = [0, 5] # Initial values of y and y' respectively (initial => t = 0).
timePoints = [5, 10, 15] # A selected list of time points.
dependents = odeint(dy_dx, initialValues, timePoints)
"""
'dependents' contains an array of lists, and in each list:
index = 0 has the y value for a value of t
index = 1 has the y' value for a value of t
"""
dependents
```

```python
y = dependents[:, 0] # Picking only the y values for the given t values.
plt.scatter(timePoints, y)
plt.title("Solution of $y''+3y'-5y=0$")
plt.xlabel("t")
plt.ylabel("y")
None
```
