# Logistical model

$\frac{dP}{dt} = kP\frac{N - P}{N}$

N is carrying capacity<br>
P is population size<br>
k is per capita growth<br><br>
Hence, N - P is the number of indivuals that can still be added<br>
$\frac{N - P}{N}$ is the fraction of unused capacity<br>
Growth rate declines with reducing capacity

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

Biologists stock a tank with 500 fishes and estimate the carrying capacity to be 10000. The number of fishes increased by 3 times during the first year. Assuming that the fish population growth satisfies the logistic equation, find population after t years.

```python
def h(P, t):
    eq = k * P * (1 - P/10000) # equation for logistic model where we take N = 10000 
    return eq
k = 3 # per capita growth rate
P0 = 500 # population at time t = 0  (the initial condition)
t = np.linspace(0, 20, 10) # time points
growth = odeint(h, P0, t) # odeint(DE, Intial Condition, time points)
plt.plot(t, growth, marker="o")
None
```

Biologists stock 50 bacterias and estimate the carrying capacity to be 1000. Assuming that the growth of bacteria satisfies the logistic equation, find population after t years. For the following growth rates: a) 0.39 b) 0.535 c) 0.409 d) 0.93

```python
def h(P, t):
    eq = k * P * (1 - P/1000) # equation for logistic model where we take N = 700 
    return eq

k1 = 0.39
k2 = 0.535
k3 = 0.409
k4 = 0.93

P0 = 18 # population at time t = 0
t = np.linspace(0, 20, 20) # time points

k = k1
growth1 = odeint(h, P0, t)

k = k2
growth2 = odeint(h, P0, t)

k = k3
growth3 = odeint(h, P0, t)

plt.plot(t, growth1, color = "red")
plt.plot(t, growth2, color = "green")
plt.plot(t, growth3, color = "blue")
plt.legend(["k1", "k2", "k3"])

None
```
