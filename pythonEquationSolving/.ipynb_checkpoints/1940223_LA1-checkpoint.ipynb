```python
import numpy as np
from scipy.integrate import odeint
from sympy import *
from sympy.interactive import printing
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

# Solve the following system of equations:
<br>
$x + y + z = 5; 2x + 3y + 5z = 8; 4x + 5z = 2$

```python
A = np.matrix([[1, 1, 1], [2, 3, 5], [4, 0, 5]])
B = np.matrix([[5], [8], [2]])
X = np.linalg.solve(A, B)
print(X)
print("x = {0}".format(float(X[0][0])))
print("y = {0}".format(float(X[1][0])))
print("z = {0}".format(float(X[2][0])))
```

# Solve the ODE:

$\frac{dy}{dx} = 5 + y; y(0) = 2$

```python
def dy_dt(y, x):
    return 5 + y
y0 = 2
x = np.linspace(-30, 30, 100)
y = odeint(dy_dt, y0, x)
plt.plot(x, y)
plt.title("Question 1.1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Plot the following

## Vector field for $f(z) = 2i - 3j$

```python
ax = axes(projection = '3d')
i, j, z = meshgrid(
    np.linspace(-10, 10, 10),
    np.linspace(-10, 10, 10),
    np.linspace(-10, 10, 10))
u = i
v = j
w = 2*i - 3*j

ax.quiver(i, j, z, u, v, w, length = 0.05)
title("f(z) = 2i - 3j", fontsize = 40)
ax.set_xlabel("i", fontsize = 30)
ax.set_ylabel("j", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Scatter plot for an ellipse

```python
ax = axes(projection = '3d')
# These values are optional, and variable...
u = 0    # x-position from origin
v = 0    # y-position from origin
a = 1    # x-axis radius
b = 2    # y-axis radius

# Plotting the eclipse
t = np.linspace(0, 2 * np.pi, 20)
x = v + b * np.sin(t)
y = u + a * np.cos(t)
z = 0
ax.scatter3D(x, y, z)

# Formatting
title("Eclipse scatterplot", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## A surface plot for $z = sin(x) + 2sin(y)$

```python
ax = axes(projection = '3d')
x, y = meshgrid(
    arange(1, 11, 0.1),
    arange(1, 11, 0.1))

z = np.sin(x) + 2*np.sin(y) # Appropriate 2D array formed
ax.plot_surface(x, y, z, color = "magenta")
title("z = sin(x) + 2sin(y)", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## The graphs of the trigonometric functions $cos(x), cos(x-2), 3cos(x+2)$

```python
xvalues = np.linspace(-np.pi, np.pi)
yvalues1 = np.cos(xvalues)
yvalues2 = np.cos(xvalues - 2)
yvalues3 = 3 * np.cos(xvalues + 2)

plt.plot(xvalues, yvalues1, lw=2, color='red', label='sin(x)')
plt.plot(xvalues, yvalues2, lw=2, color='blue', label='2sin(x)')
plt.plot(xvalues, yvalues3, lw=2, color='green', label='sin(2x)')

plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.axhline(lw=0.5, color='purple')
plt.axvline(lw=0.5, color='purple')
plt.legend()

None
```

# Using two square matrices A and B, show that $(A + B)^T = A^T + B^T$

```python
A = np.matrix([[1, 2], [3, 4]])
B = np.matrix([[2, 1], [1, 3]])
if all(np.transpose(A + B) == np.transpose(A) + np.transpose(B)):
    print("(A + B)^T = A^T + B^T")
else:
    print("(A + B)^T ≠ A^T + B^T")
```

# Plot the slope field for the differential equation, $\frac{dy}{dx} = 3sin(x) - e^x$

```python
x, y = meshgrid(
    np.linspace(-2*np.pi, 2*np.pi, 20),
    np.linspace(-1, 1, 20))

dy = 3*np.sin(x) - np.e**x
dx = np.ones(dy.shape)

plt.quiver(x, y, dx, dy, color = "blue")
plt.title("dy/dx = 3sin(x) - e^x", fontsize = 40)
plt.xlabel("x", fontsize = 30)
plt.ylabel("y", fontsize = 30)
None
```

# Obtain stream plot for the diffrential equation, $\frac{dy}{dx} = 2tan(x) + xy$

```python
x, y = meshgrid(
    np.linspace(-2*np.pi, 2*np.pi, 20),
    np.linspace(-1, 1, 20))

dy = 2*np.tan(x) + x*y
dx = np.ones(dy.shape)

plt.streamplot(x, y, dx, dy, color = "purple")
plt.title("dy/dx = 2tan(x) + xy")
None
```
