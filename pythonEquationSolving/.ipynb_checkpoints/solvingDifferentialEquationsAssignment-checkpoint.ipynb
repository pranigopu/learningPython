```python
from numpy import *
from scipy.integrate import odeint
from sympy import *
from sympy.interactive import printing
from matplotlib.pyplot import *
```

# Solve the following ODE using odeint( ) function

## $\frac{dy}{dt} = 2 + y^2; y(0) = 5$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t) - 2 - y**2, 0) 
gs = dsolve(B, y)

aftersubs = gs.subs({t:0, y:5})
C1_value = solve(aftersubs)
print("C1 =", C1_value)

ps = gs.subs(C1, -atan(5*sqrt(2)/2))
ps
```

```python
def dy_dt(y, t):
    return 2 + y**2
y0 = 5
t = linspace(-30, 30, 100)
y = odeint(dy_dt, y0, t)
plot(t, y)
title("Question 1.1")
xlabel("t")
ylabel("y")
show()
```

## $\frac{dy}{dt} + 5y + 2t; y(0) = 0$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t) + 5*y + 2*t, 0) 
gs = dsolve(B, y)

aftersubs = gs.subs({t:0, y:0})
C1_value = solve(aftersubs)
print("C1 =", C1_value)

ps = gs.subs(C1, -2/25)
ps
```

```python
def dy_dt(y, t):
    return -(5*y + 2*t)
y0 = 0
t = linspace(-30, 30, 100)
y = odeint(dy_dt, y0, t)
plot(t, y)
title("Question 1.2")
xlabel("t")
ylabel("y")
show()
```

## $\frac{dy}{dt} = sin(t); y(0) = 5$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t) - sin(t), 0) 
gs = dsolve(B, y)

aftersubs = gs.subs({t:0, y:5})
C1_value = solve(aftersubs)
print("C1 =", C1_value)

ps = gs.subs(C1, 6)
ps
```

```python
def dy_dt(y, t):
    return sin(t)
y0 = 5
t = linspace(-30, 30, 100)
y = odeint(dy_dt, y0, t)
plot(t, y)
title("Question 1.2")
xlabel("t")
ylabel("y")
show()
```

# Obtain the solution curves for the ODE $\frac{dp}{dt} = e^t$

```python
def dp_dt(p, t):
    return e**t
t = linspace(0, 5, 100)
for i in range(0, 51):
    p0 = i
    p = odeint(dp_dt, p0, t)
    plot(p, t)

title("Question 2")
xlabel("t")
ylabel("p")
show()
```

# Solve the differential equations

## $\frac{dy}{dt} + 5y = 0$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t) + 5*y, 0) 
gs = dsolve(B, y)
gs
```

## $\frac{dy}{dt} = y - 2t$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t), y - 2*t) 
gs = dsolve(B, y)
gs
```

## $\frac{d^2y}{dt^2} + 5y = 0$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t, t) + 5*y, 0) 
gs = dsolve(B, y)
gs
```

## $\frac{d^3y}{dt^3} + 10\frac{d^2y}{dt^2} + \frac{dy}{dt} = sin(t)$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t, t, t) + 10*y.diff(t, t) + y.diff(t), sin(t)) 
gs = dsolve(B, y)
gs
```

## $\frac{d^2y}{dt^2} + 6\frac{dy}{dt} - 9y = e^tsin(t)$

```python
printing.init_printing(use_latex = True)
t = Symbol("t") 
y = Function("y")(t)
C1 = Symbol("C1")
B = Eq(y.diff(t, t) + 6*y.diff(t) - 6*y, (e**t) * sin(t)) 
gs = dsolve(B, y)
gs
```
