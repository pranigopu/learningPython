# Solving differential equations

```python
from sympy import *
```

## $\frac{dy}{dx} = 5$

```python
# Defining the independent variable
x = Symbol("x") # declare some variables for the use of mathematical expression 
# Defining the dependent variable
y = Function("y")(x)  # declaring y as a function of x
# Assigning differential equation
# The differential equation format is Eq(LHS, RHS)
A = Eq(y.diff(x), 5)
# To solve the given ODE
# Syntax: dsolve(equation, function)
solution = dsolve(A, y)
print("The solution of the given differential equation is", solution)
```

## $6y + \frac{d^2y}{dx^2} + 5\frac{dy}{dx} = 0$

```python
printing.init_printing(use_latex = True)
x = Symbol("x") 
y = Function("y")(x) 
C = Eq(6*y + y.diff(x, x) + 5*y.diff(x), 0) 
dsolve(C, y)
```

## (Particular solution) $\frac{df}{dx} = 5f ; f(2) = 10$

```python
printing.init_printing(use_latex = True)
x = Symbol("x") 
f = Function("f")(x)
C1 = Symbol("C1")
D = Eq(f.diff(x) - 5*f, 0)

general_solution = dsolve(D, f)

after_substituting = general_solution.subs({x:2, f:10})
# subs accepts a dictionary as argument, that maps the symbol to a value
C1_value = solve(after_substituting)
print("C1 =", C1_value)

particular_solution = general_solution.subs(C1, 10*exp(-10))
particular_solution
```

## Finding the numerical solution of an ODE

### Solve $y' = 1; y(0) = 0$

We find the numerical solution of an ODE using the package Scipy.integrate. The odeint function returns the y-values from the solution of the differential equation.
<br><br>
odeint has the following syntax: odeint(functionReturningDifferentialEquation, initialCondition, timePoints)

```python
from numpy import*
from matplotlib.pyplot import *
from scipy.integrate import odeint
```

```python
# Function that returns dy/dt
# Defining a function is important,
# since the odeint function accepts function name
# as an argument for finding the derivative.
def dy_dt(y, t):
    return 1

# The initial condition
y0 = 0

# Time points
t = linspace(0, 30)

# Solve ODE
y = odeint(dy_dt, y0, t)

# Plot result
plot(t, y)
show()
```

### Solve $\frac{dy}{dt} = t; y(0) = 0.5$

```python
# Function that returns dy/dt
def dy_dt(y, t):
    return t

# The initial condition
y0 = 0.5

# Time points
t = linspace(0, 30)

# Solve ODE
y = odeint(dy_dt, y0, t)

# Plot result
plot(t, y)
show()
```

### Solve $y' = x - y; y(0) = 1$

```python
# Function that returns dy/dt
def dy_dx(y, x):
    return  x - y

# The initial condition
y0 = 1

# Time points
x = linspace(0, 30)

# Solve ODE
y = odeint(dy_dx, y0, x)

# Plot result
plot(x, y, color = "green")
title("Gud")
xlabel("x")
ylabel("y")
```

### Solve $y' = -ky; y(0) = 5; k = \{0.1, 0.2, 0.5\}$

#### METHOD 1

```python
# Function that returns dy/dt
def dy_dx_1(y, x):
    return  -0.1*y
def dy_dx_2(y, x):
    return  -0.2*y
def dy_dx_3(y, x):
    return  -0.5*y

# The initial condition
y0 = 5

# Time points
x = linspace(0, 30)

# Solve ODE for different k values
y1 = odeint(dy_dx_1, y0, x)
y2 = odeint(dy_dx_2, y0, x)
y3 = odeint(dy_dx_3, y0, x)

# Plot result
plot(x, y1, '--', color = "green")
plot(x, y2, '.', color = "red")
plot(x, y3, ':', color = "blue")
title("Noice")
xlabel("x")
ylabel("y")
legend(["k = 0.1", "k = 0.2", "k = 0.5"])
show
```

#### METHOD 2

```python
# Function that returns dy/dt
def dy_dx_1(y, x):
    return  -k*y

# The initial condition
y0 = 5

# Time points
x = linspace(0, 30)

# Solve ODE for different k values
k = 0.1
y1 = odeint(dy_dx_1, y0, x)
k = 0.2
y2 = odeint(dy_dx_2, y0, x)
k = 0.3
y3 = odeint(dy_dx_3, y0, x)

# Plot result
plot(x, y1, '--', color = "green")
plot(x, y2, '.', color = "red")
plot(x, y3, ':', color = "blue")
title("Noice")
xlabel("x")
ylabel("y")
legend(["k = 0.1", "k = 0.2", "k = 0.5"])
show
```

### Solution curves for $y' = x$

```python
# Function that returns dy/dt
def dy_dx(y, x):
    return  x

# Time points
x = linspace(0, 5)

for i in range(0, 11):
    # The initial condition
    y0 = i
    
    # Plotting the current curve
    y = odeint(dy_dx, y0, x)
    plot(x, y)

title("Hmmm")
xlabel("x")
ylabel("y")
show()
```

### Solution curves for $y' = cos(x) + 1$

```python
# Function that returns dy/dt
def dy_dx(y, x):
    return  cos(x) + 1

# Time points
x = linspace(0, 20)

for i in range(0, 11):
    # The initial condition
    y0 = i
    
    # Plotting the current curve
    y = odeint(dy_dx, y0, x)
    plot(x, y)

title("Oooh")
xlabel("x")
ylabel("y")
show()
```
