# AIM

Linear transformation is a homomrphism between two vector spaces. It is a one-to-one function from a domain to a range, and hence, can be plotted. This is the focus of this record.

Let U and V be two vector spaces of two dimensional vectors defined over the field of real numbers. Consider the transformation $T:U-->V$ such that $T(u)=T((u_1,u_2))=(u_1+u_2,2u_1-u_2)$

```python
import matplotlib.pyplot as plt
import numpy as np
u1, u2 = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
v1, v2 = u1 + u2, 2*u1 - u2
plt.quiver(u1, u2, v1, v2, alpha = 1)
plt.title("\n$T(u)=T((u_1,u_2))=(u_1+u_2,2u_1-u_2)$\n", size = 20)
plt.xlabel('$u_1$', size = 20)
plt.ylabel('$u_2$', size = 20)
None
```

# CONCLUSION

Here, each set of coordinates $(u_1,u_2)$ is a vector from the vector space U, and the vector originating from this point on the graph is the transformed vector.
