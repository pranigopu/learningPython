# PLOTTING IN 3D

```python
from matplotlib.pyplot import *
from mpl_toolkits import mplot3d
from numpy import *
# Due to * we need not refer to the module name when we need to use a function or class
```

## 3D plots

```python
ax = axes(projection = '3d')
t = linspace(0, 1, 10) # The arguments are: lower bound, upper bound, step size... it creates an array accordingly
xval = -2*sqrt(t) + 4*t**2
yval = 8*t**2
zval = 7*t**3
ax.plot3D(xval, yval, zval, 'red')
title("What is this???", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
show()
```

```python
ax = axes(projection = '3d')
xval = linspace(0, 16, 1000)
yval = sin(xval)
zval = cos(xval)
ax.plot3D(xval, yval, zval, color = "magenta")
title("Helix", fontsize = 20, weight = "bold", style = "italic")
ax.set_xlabel("x", weight = "bold")
ax.set_ylabel("y", weight = "bold")
ax.set_zlabel("z", weight = "bold")
show()
```

## Scatter plot

```python
ax = axes(projection = '3d')
z = linspace(0, 100, 1000)
y = sin(z)
x = cos(z)
ax.scatter3D(x, y, z)
show()
```

```python
ax = axes(projection = '3d')
z = 30 * random.random(100)
# This random function call generates an array of 100 random values less than 1
y = sin(z)
x = cos(z)
ax.scatter3D(x, y, z)
show()
```

## Vector fields

A vector field is an assignment of a vector to each point in a subset of space.<br>
For instance, a vector field in the plane can be visualised as a collection of arrows with a given magnitude and direction, each attached to a point in the plane.

```python
ax = axes(projection = '3d')
x, y, z = meshgrid(
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2))
u = sin(x) + cos(y) + tan(z)
v = sin(x) * cos(y) * tan(z)
w = sin(x) - cos(y) * tan(z)
ax.quiver(x, y, z, u, v, w, length = 0.1)
show()
```
