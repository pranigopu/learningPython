# Slope fields and stream plots

A slope field is a collection of short line segments, whose slopes match that of a solution of a first-order differential equation passing through the segment's midpoint. The pattern produced by the slope field aids in visualizing the shape of the curve of the solution.

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Slope fields

$\frac{dy}{dx} = sin(x) âˆ’ 2 cos(y)$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = np.sin(X) - 2*np.cos(Y)
dx = np.ones(dy.shape)

plt.quiver(X, Y, dx, dy, color = "blue")
plt.title("dy/dx = sin(x) - 2cos(y)")
None
```

$\frac{dy}{dx} = cos(3xy)$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = np.cos(3*X*Y)
dx = np.ones(dy.shape)

plt.quiver(X, Y, dx, dy, color = "red")
plt.title("dy/dx = cos(3xy))")
None
```

$\frac{dy}{dx} = x + 2xy âˆ’ y$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = X + 2*X*Y - Y
dx = np.ones(dy.shape)

plt.quiver(X, Y, dx, dy, color = "purple")
plt.title("dy/dx = + x + 2xy - y")
None
```

## Stream plots

$\frac{dy}{dx} = tan(x) + xy$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = np.tan(X) + X*Y
dx = np.ones(dy.shape)

plt.streamplot(X, Y, dx, dy, color = "purple")
plt.title("dy/dx = tan(x) + xy")
None
```

$\frac{dy}{dx} = 3x âˆ’ sec(y)$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = 3*X + 1/np.cos(Y)
dx = np.ones(dy.shape)

plt.streamplot(X, Y, dx, dy, color = "purple")
plt.title("dy/dx = 3x + sec(y)")
None
```

$\frac{dy}{dx} = \frac{x}{y}$

```python
# Making a set of x and y coordinates to define the intervals on each axis...
x = np.linspace(-2*np.pi, 2*np.pi, 20)
y = np.linspace(-1, 1, 20)

# Making the set of all points corresponding to the given set of x and y coordinates.
# This creates a set of x and y values for each point, which can be used in the differential equation
X, Y = np.meshgrid(x, y)

dy = X/Y
dx = np.ones(dy.shape)

plt.streamplot(X, Y, dx, dy, color = "purple")
plt.title("dy/dx = x/y")
None
```
