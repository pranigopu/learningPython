## Relative growth using exponential model

```python
from sympy import Symbol, Function, Eq, dsolve, solve, exp
from numpy import log
t = Symbol('t') # Number of years since the current time
p = Function('p')(t) # Population after t years
k = Symbol('k') # Relative growth rate
p0 = 20 # Initial population
growthModel = Eq(p, p0*exp(k*t))
growthModel
```

At t = 10, p = 55. We use this value since it is the next available data point in the data set, offering us a clearer picture of the relative growth rate. Hence, at t = 100, we get...

```python
eq = growthModel.subs({t:10, p:55})
solve(eq, k)
```

Hence we get $*log(11^{1/10})\frac{2^{4/5}}{2}$ as the only real solution of k. Hence, the relative growth rate is...

```python
 log(11**(1/10)*2**(4/5)/2)
```

## Logistic growth model vs. exponential growth model

$\frac{dP}{dt} = kP\frac{N - P}{N}$

N is carrying capacity, <br>
P is population size, <br>
k is per capita growth. <br><br>
Hence, N - P is the number of indivuals that can still be added. <br>
$\frac{N - P}{N}$ is the fraction of unused capacity. <br>
Growth rate declines with reducing capacity

```python
from scipy.integrate import odeint
```

```python
# LOGISTIC GROWTH
def Population(P_log, T):
    eq = k * P_log * (1 - P_log/300) # Equation for logistic model where we take N = 700 
    return eq
k = 0.101160091167848 # Per capita growth rate (taken from previous calculation)
P0 = 20 # population at time t = 0  (the initial condition)
P_log = odeint(Population, P0, T) # odeint(DE, Intial Condition, time points)

# EXPONENTIAL GROWTH
P_exp = 20*float(exp(0.101160091167848))**T

#PLOTS
plot(T, P_log)
plot(T, P_exp)
plot(T, P, color = "magenta")
legend(["Logistic growth", "Exponential growth", "Original"])
title("Growth models vs. actual data", size = 20)
xlabel("Time (in years)", size = 15)
ylabel("Population (in lakhs)", size = 15)
axis([0, 100, 0, 400])
None
```

# Radioactive decay

We use the formula $N = \frac{N_0}{2^n}$

where $N$ is the mass of the substance, $N_0$ is the initial mass of the substance, and $n$ is the number of half lives passed.

```python
n = Symbol('n') # Number of half lives passed
N = Function('N')(n) # Mass of substance
N0 = 15 # Initial mass of substance
decay = Eq(N, N0/2**n)
decay
```

If half live is 16 hours, 24 hours is 24/16 half lives.

```python
decay.subs({n:24/16})
```

Hence, the mass of the substance after 24 hours will be around 5.30330085889911 grams.
