# Assignment 2: solving linear equation systems

```python
import numpy as np
def getCoefficient(equation, variable):
    i = 0
    n = 0
    power = 1
    length = len(equation)
    
    while i < length and not(equation[i] == variable):
        i = i + 1
    if i >= len(equation):
        return 0
    
    i = i - 1
    if i < 0 or equation[i] == ' ' or equation[i] == '\t':
        return 1
    while equation[i] >= '0' and equation[i] <= '9' and i > -1:
        n = n + int(equation[i]) * power
        power = power * 10
        i = i - 1
    return n

def solveEquations(equations):
    A = np.zeros((3, 3))
    B = np.zeros((3, 1))
    variables = ['x', 'y', 'z']
    
    for i in range(0, 3):
        # Equation i
        for j in range(0, 3):
            A[i][j] = getCoefficient(equations[i], variables[j])
        B[i][0] = getCoefficient(equations[i], ';')
        
    print("Coefficient matrix:\n{0}".format(A))
    print("Constant sum matrix:\n{0}".format(B))
    
    X = np.linalg.solve(A, B)
    print("x = {0}".format(float(X[0][0])))
    print("y = {0}".format(float(X[1][0])))
    print("z = {0}".format(float(X[2][0])))
```

```python
print("----------------\nProblem 1")
solveEquations(["2x + 7y + z = 40;", "3x + y + 4z = 15;", "x + 5y + 3z = 30;"])
print("----------------\nProblem 2")
solveEquations(["3x + 2y + 7z = 8;", "4x + 2y + 5z = 9;", "2x + 5y + 2z = 12;"])
print("----------------\nProblem 3")
solveEquations(["7x + y + z = 15;", "2xy + z = 4;", "3x + 2y + 4z = 29;"])
print("----------------\nProblem 4")
solveEquations(["2x + y + 6z = 7;", "4x + 3y + z = 8;", "x + 3y + 5z = 9;"])
print("----------------\nProblem 5")
solveEquations(["2x + y + 5z = 26;", "x + y + z = 7;", "3x + 5y + 2z = 22;"])
print("----------------\nProblem 6")
solveEquations(["7x − y + z = 4;", "5x − 2y + 2z = −1;", "3x + 2y + z = 6;"])
```
