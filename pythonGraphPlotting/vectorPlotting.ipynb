Quiver plot is a type of 2D plot which shows vector lines as arrows.

# $f(x, y) = 2(x + y) + 4(y - x)^2$

```python
from matplotlib.pyplot import axes, quiver, show, title, xlabel, ylabel
from numpy import arange, meshgrid
```

```python
x, y = meshgrid(
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2))
u = 2*(x + y)
v = 4*(y - x)**2
quiver(x, y, u, v, alpha = 0.5)
title("$f(x, y)=2(x+y)+4(y-x)^2$", size = 20)
xlabel("x", size = 15)
ylabel("y", size = 15)
# The alpha option defines the transparency of the vectors
show()
```
