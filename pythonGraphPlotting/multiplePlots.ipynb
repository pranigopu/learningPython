# Multiple plots in a single graph

```python
from pylab import plot, legend, show
```

```python
year_2001 = [25, 25, 26, 27, 28, 30, 34, 19, 15, 15, 14, 14]
year_2002 = [21, 30, 31, 31, 34, 35, 37, 27, 20, 20, 14, 12]
year_2003 = [19, 19, 23, 30, 30, 30, 29, 18, 15, 14, 12, 10]
```

Note that the domain of the data must be given before the data. The data values form the y-values, and the specified domains form the x-values

```python
plot(range(1, 13), year_2001, range(1, 13), year_2002, range(1, 13), year_2003)
show()
```

## Legend function

```python
plot(range(1, 13), year_2001, range(1, 13), year_2002, range(1, 13), year_2003)
legend([2001, 2002, 2003])
show()
```

We can specify where to show the legend in the graph...

```python
plot(range(1, 13), year_2001, range(1, 13), year_2002, range(1, 13), year_2003)
legend([2001, 2002, 2003], bbox_to_anchor = (0.8, 0.4))
show()
```
