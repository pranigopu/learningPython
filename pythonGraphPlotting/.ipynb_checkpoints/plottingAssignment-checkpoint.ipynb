```python
from matplotlib.pyplot import *
from numpy import *
from mpl_toolkits import mplot3d
```

## Question 1

```python
ax = axes(projection = '3d')
i, j, z = meshgrid(
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2))
u = i
v = j
w = i - j
ax.quiver(i, j, z, u, v, w, length = 0.1)

title("f(z) = i - j", fontsize = 40)
ax.set_xlabel("i", fontsize = 30)
ax.set_ylabel("j", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 2

```python
ax = axes(projection = '3d')
i, j, z = meshgrid(
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2),
    arange(-0.8, 1, 0.2))
u = i
v = j
w = 2*i + 3*j
ax.quiver(i, j, z, u, v, w, length = 0.1)

title("f(z) = 2i + 3j", fontsize = 40)
ax.set_xlabel("i", fontsize = 30)
ax.set_ylabel("j", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 3

```python
ax = axes(projection = '3d')
# These values are optional, and variable...
u = 100  # x-position from origin
v = 5    # y-position from origin
a = 1    # x-axis radius
b = 10   # y-axis radius

# Plotting the eclipse
t = linspace(0, 2 * np.pi, 20)
x = 7
y = u + a * cos(t)
z = v + b * sin(t)
ax.scatter3D(x, y, z)

# Formatting
title("Eclipse scatterplot", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 4

```python
ax = axes(projection = '3d')
# Forming the set of x and y value in one function (note: upper = 11 => there will be 10 elements)
x, y = meshgrid(
    arange(1, 11, 1),
    arange(1, 11, 1))

z = 3 * ones((10, 10)) # 2D array of 3s to match x's and y's
ax.plot_surface(x, y, z)
title("z = 3", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 5

```python
# Forming the set of x and y value in one function (note: upper = 11 => there will be 10 elements)
ax = axes(projection = '3d')
x, z = meshgrid(
    arange(1, 11, 1),
    arange(1, 11, 1))

y = ones((10, 10)) # 2D array of 1s
ax.plot_surface(x, y, z)
title("y = 1", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 6

```python
# Forming the set of x and y value in one function (note: upper = 11 => there will be 10 elements)
ax = axes(projection = '3d')
y, z = meshgrid(
    arange(1, 11, 1),
    arange(1, 11, 1))

x = 2 * ones((10, 10)) # 2D array of 2s
ax.plot_surface(x, y, z)
title("x = 2", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 7

```python
ax = axes(projection = '3d')
# Same principle as the last 3 questions except that no axis is constant
x, y = meshgrid(
    arange(1, 11, 0.1),
    arange(1, 11, 0.1))

z = cos(x) + cos(y) # Appropriate 2D array formed
ax.plot_surface(x, y, z, color = "magenta")
title("z = cos(x) + cos(y)", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 8

```python
ax = axes(projection = '3d')
# Same principle as the last question
x, y = meshgrid(
    arange(1, 2, 0.1),
    arange(1, 2, 0.1))
z = tan(x) + tan(y) # Appropriate 2D array formed
ax.plot_surface(x, y, z, color = "red")
title("z = tan(x) + tan(y)", fontsize = 40)
ax.set_xlabel("x", fontsize = 30)
ax.set_ylabel("y", fontsize = 30)
ax.set_zlabel("z", fontsize = 30)
None
```

## Question 9

```python
x = linspace(-2, 2, 100)

y = cos(x)
plot(x, y, color = "green")

y = cos(2*x)
plot(x, y, color = "red")

y = cos(3*y)
plot(x, y, color = "blue")

title("3 in 1 trigonometric graphs", fontsize = 25)
xlabel("x", fontsize = 20)
ylabel("y", fontsize = 20)
grid()
None
```
