```python
from pylab import plot, show
```

```python
x_coords = [1, 5, 9]
y_coords = [2, 10, 7]
```

```python
plot(x_coords, y_coords)
```

```python
plot(x_coords, y_coords)
show()
```

```python
plot(x_coords, y_coords, marker = "o")
show()
```
