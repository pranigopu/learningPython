# Multiple plots in a single graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvalues = np.linspace(-np.pi, np.pi)
yvalues1 = np.sin(xvalues)
yvalues2 = 2 * np.sin(xvalues)
yvalues3 = np.sin(2 * xvalues)

plt.plot(xvalues, yvalues1, lw=2, color='red', label='sin(x)')
plt.plot(xvalues, yvalues2, lw=2, color='blue', label='2sin(x)')
plt.plot(xvalues, yvalues3, lw=2, color='green', label='sin(2x)')

plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('sin(x)')

# To draw axis lines
plt.axhline(lw=0.5, color='purple')
plt.axvline(lw=0.5, color='purple')
plt.legend()

None
```

# Subplots

NOTES
<br>
matplotlib.pyplot.subplot(n, m, i)
<br><br>
This function is placed above the plotting function for the desired subplot. The plotting function is as usual.
<br><br>
The first 2 arguments define the dimensions of the grid in which the subplots will be placed. Here, we have defined a n x m grid, meaning nm subplots in a n x m square arangement.
<br><br>
The last argument is the index of the current plot. In this case, this plot is the ith plot in the n x m grid of subplots.

```python
xvalues = np.linspace(-np.pi, np.pi) # Creates a vector from the defined range
plt.suptitle('Polynomial Functions') # Title over all subplots

plt.subplot(2, 2, 1)
yvalues = xvalues**1
plt.plot(xvalues, yvalues, lw=2, color='red', label='y = x')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
yvalues = xvalues**2
plt.plot(xvalues, yvalues, lw=2, color='blue', label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 3)
yvalues = xvalues**3
plt.plot(xvalues, yvalues, lw=2, color='green', label='y = x^3')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 4)
yvalues = xvalues**4
plt.plot(xvalues, yvalues, lw=2, color='purple', label='y = x^4')
plt.xlabel('x')
plt.ylabel('y')

# The following axis drawing functions only apply for the just preceding plot.
plt.axhline(lw=0.5, color='purple')
plt.axvline(lw=0.5, color='purple')
plt.legend()

plt.tight_layout()
# Must be given after the subplots. Ensures that there is sufficient space between the subplots

None
```
