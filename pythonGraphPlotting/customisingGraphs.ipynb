# Customising graphs

## Labelling and marking

```python
from pylab import plot, show, legend, title, xlabel, ylabel, axis
```

```python
data1 = [1, 2, 3, 4, 5, 6, 7, 8, 7, 6, 5, 4]
data2 = [2, 4, 8, 16, 32, 64, 32, 16, 8, 4, 2, 1]
data3 = [1, 2, 5, 7, 12, 19, 31, 19, 12, 7, 5, 3]
```

```python
plot(range(1, 13), data1, range(1, 13), data2, range(1, 13), data3, marker = 'o')
legend([2001, 2002, 2003])
title("Average temperatures in Lalaland")
xlabel("Month")
ylabel("Temperature")
show()
```

## Axes

We can set the minimum of the x or y axes as follows...

```python
datax = [1, 2, 4, 8, 12, 14, 15, 15.5, 15, 14, 12, 8, 4, 2, 1]
```

```python
plot(range(1, 16), datax)
show()
```

```python
plot(range(1, 16), datax)
axis([1, 8, 0, 12])
show()
```

(This can allow you to effectively zoom in and out of particular areas of the graph)
